```python
govt_data = pd.read_csv("govtData.csv")
```

```python
govt_data.head()
```

```python
govt_data = govt_data.set_index("Affected States/UTs")
```

```python
govt_data.T["C/D"].values
```

```python
causeOrDeath = govt_data.T["C/D"].values
govt_data = govt_data[(govt_data["2010"] !="C").values]
```

```python
govt_data.columns = pd.MultiIndex.from_tuples(zip(govt_data.columns, causeOrDeath))
```

```python
govt_data.head()
```

```python
a =govt_data.columns.levels[0].get_values()
```

```python
a = pd.Series(a).str.replace(r'\.1', '').str.replace(r'\D+', '')
```

```python
govt_data.columns = pd.MultiIndex.from_tuples(zip(a.get_values(), causeOrDeath))
```

```python
govt_data.head()
```

```python
govt_data = govt_data[govt_data.columns].astype(int)
```

```python
govt_data.columns = govt_data.columns.swaplevel(0, 1)
govt_data.sortlevel(0, axis=1, inplace=True)
govt_data.head()
```
